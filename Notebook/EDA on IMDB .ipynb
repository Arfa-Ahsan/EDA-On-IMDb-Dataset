# Exploratory Data Analysis on IMDB Dataset

## Problem Statement

##### The goal of this project is to analyze the imdb dataset and then present meaningful insights to better understand the  trends,relationships, and patterns found in the data.

## Data Description

#### The following is the description of the data;
- Poster_Link: URL link to the movie's poster image.

- Series_Title: Title of the movie or series.

- Released_Year: Year the movie was released.

- Certificate: Certification or rating given to the movie (e.g., PG, R).

- Runtime: Duration of the movie in minutes.

- Genre: Genre(s) of the movie (e.g., Action, Comedy).

- IMDB_Rating: Rating of the movie on IMDb, typically on a scale from 1 to 10.

- Overview: Brief summary or synopsis of the movie's plot.

- Meta_score: Score given by Metacritic, reflecting critics' reviews (typically on a scale from 0 to 100).

- Director: Name of the movie's director.

- Star1: Name of the first lead actor or actress.

- Star2: Name of the second lead actor or actress.

- Star3: Name of the third lead actor or actress.

- Star4: Name of the fourth lead actor or actress.

- No_of_Votes: Number of votes the movie has received on IMDb.

- Gross: Total revenue earned by the movie, usually in millions of dollars.

## Data Cleaning

```python
#check for duplicated rows
duplicated_rows=df.duplicated().sum()
print("The number of duplicated rows are:",duplicated_rows)
```

#### There are no duplicates in the dataset.

```python
#removing unnecessary columns
df=df.drop(columns=["Poster_Link","Overview"])
df.head()
```

#### Removing both of these columns as they dont provide any meaningful insights

```python
#Extracting the min from the runtime column and changing its data type
df["Runtime"]=df["Runtime"].str.strip("min").astype('int64')
df.head()
```

```python
#verifying the change
assert df['Runtime'].dtype=='int64', 'The Runtime column has invalid data type'
```

```python
# Extracting the comma from gross column and changing its datatype
df["Gross"]=df["Gross"].str.replace(",","").astype('float64')
df.head()
```

```python
#verifying the change
assert df['Gross'].dtype=='float64', 'The Gross column has invalid data type'
```

```python
# Checking if the released year contain values other than digits
try:
    if not df['Released_Year'].str.isdigit().all():
         raise ValueError('The released year column contain values other than digits')
except ValueError as t:
    print(t)
```

```python
#finding out the value that is invalid in released year column
df = df.copy()
df[~df['Released_Year'].str.isdigit()]
```

#### The released year has a value PG which is likely due to an error.

```python
# filtering the rows so the invalid value can be removed
df=df[df['Released_Year'].str.isdigit()]
```

```python
# verifying the change
assert df['Released_Year'].str.isdigit().all(), 'The released year column contain values other than digits'
```

```python
#changing the datatype of released year
df['Released_Year'] = df['Released_Year'].astype('int64')

# verifying the change
assert df['Released_Year'].dtype=='int64', 'The released year column contain invalid data type'
```

```python
#checks for the null values 
missing_values=df.isna().sum()
print(missing_values)
```

#### The column certificate,meta score and gross have missing values.

```python
# Finding out the percentage of missing values in the dataset.
missing_values_percent=df.isna().sum()/len(df)*100
print(missing_values_percent)
```

```python
#handling the missing data 
num_cols=df.select_dtypes(include=["int64","float64"]).columns

for cols in num_cols:
     df[cols]=df[cols].fillna(df[cols].median())
        
df.isna().sum()
```

####  Imputing missing values of numerical columns with the median.

```python
cat_cols=df.select_dtypes(include="object").columns

for cols in cat_cols:
     df[cols]=df[cols].fillna(df[cols].mode()[0])
        
df.isna().sum()
```

#### Imputing the missing values for  categorical columns with the most frequent value

```python
#checking for garbage values
print("============================")
for cols in cat_cols:
    print(df[cols].value_counts())
    print("=============================")
```

```python
#checking if data is out of range or not 
assert ((df["IMDB_Rating"] >= 1.0) | (df["IMDB_Rating"] <= 10.0)).all(), "IMDB Rating is out of range."
assert ((df["Meta_score"] >= 0.0) | (df["Meta_score"] <= 100.0)).all(), "Meta Score is out of range."
```

### What are the top 10 movies based on their ratings?

```python
# find the average rating for top 10 movies
highest_rated_movie=df.groupby('Series_Title')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
print(f'Top 10 Movies based on average rating :\n{highest_rated_movie}')
```

```python
plt.figure(figsize=(9,7))
sns.barplot(x=highest_rated_movie.index,y=highest_rated_movie.values,palette=colors)

#Annotating each bar
for i,v in enumerate(highest_rated_movie):
    plt.text(i,v+0.05,s=v,ha="center",fontsize=11,fontweight="bold")

#Add the titles and labels
plt.title('Top 10 highest rated movies')
plt.xlabel('Series Title')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=90)
plt.show()
```

### Insights
- 'The Shawshank Redemption' was the most liked movie with a rating of 9.3 .
- The second most liked movie by audience was 'The GodFather' with a rating of 9.2.
